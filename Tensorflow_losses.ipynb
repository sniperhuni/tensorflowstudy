```
import tensorflow as tf
```

```
'''
Binary Cross entropy
: when there are only two label classes
  For each example, there should be a single floating-point value per prediction

tf.keras.losses.BinaryCrossentropy(
  from_logits=False, label_smoothing=0, reduction=loss_utils.ReductionV2.AUTO,
  name='binary_crossentropy'
)

Cross entropy = -(y*ln(p) - (1-y)*ln(1-p))

from_logits = True ; y_pred is not normalized , outputs are not passed a softmax function

'''
```

```
# Standalone usage
y_true = [[0., 1.], [0., 0.]]
y_pred = [[0.6, 0.4], [0.4, 0.6]]

# in this case, batch_size = 4 (total number of sample)
# (-(1-0)*ln0.6 -1*ln(0.4) - (1-0)*ln0.6 - (1-0)*ln0.4)/4 = 0.815
# AUTO : 'sum_over_batch_size' reduction type

bce = tf.keras.losses.BinaryCrossentropy()
bce(y_true, y_pred).numpy()
```

```
bce(y_true, y_pred, sample_weight= [1,0]).numpy()
```

```
bce = tf.keras.losses.BinaryCrossentropy(reduction=tf.keras.losses.Reduction.NONE)
bce(y_true, y_pred, sample_weight= [1,0]).numpy()
```

```
bce = tf.keras.losses.BinaryCrossentropy(reduction=tf.keras.losses.Reduction.SUM)
bce(y_true, y_pred).numpy()
```

```
'''
CategoricalCrossentropy ; label to be provided in a one_hot encoded representation

SparseCategoricalCrossentropy ; label to be provided as integers
'''
```

```
y_true = [[0,1,0], [0,0,1]]
y_pred = [[0.05, 0.95, 0], [0.1, 0.8, 0.1]]

# (-1*ln(0.95) + (-1 * ln(0.1))) / 2 = 1.176

cce = tf.keras.losses.CategoricalCrossentropy()
cce(y_true, y_pred).numpy()
```
